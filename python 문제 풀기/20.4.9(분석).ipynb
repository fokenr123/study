```python
data2 = pd.read_csv('./data/total.csv' ,index_col = ['date'])
data2
```

```python
# 일일확진자수, 일일격리해제수, 일일사망자수 구하기

Daily_confirmed=[]
Daily_recoverd=[]
Daily_death=[]
Daily_confirmed.append(595) # 3월 1일 일일확진자수는 질병관리본부에서 가져옴
Daily_recoverd.append(3)  # 3월 1일 일일격리해제수는 질병관리본부에서 가져옴
Daily_death.append(1) # 3월 1일 일일사망자수는 질병관리본부에서 가져옴

for i in range(len(data2)-1) :
    Daily_confirmed.append(data2.iloc[:,1][i+1] - data2.iloc[:,1][i])
    Daily_recoverd.append(data2.iloc[:,2][i+1] - data2.iloc[:,2][i])
    Daily_death.append(data2.iloc[:,4][i+1] - data2.iloc[:,4][i])

a = pd.DataFrame(Daily_confirmed, index = data2.index)
a = a.rename(columns = {0:"일일확진자"})

b = pd.DataFrame(Daily_recoverd, index = data2.index)
b = b.rename(columns = {0:"일일격리해제"})

c = pd.DataFrame(Daily_death, index = data2.index)
c = c.rename(columns = {0:"일일사망"})

today = pd.concat([data2, a, b, c], axis =1)
today
```
